```python
data_dropped = Data.dropna()

```

```python
data_dropped.shape
```

```python
missing_values = data_dropped.apply(lambda x: x.isnull().sum().sum())
print(missing_values)
```

```python
data_dropped.describe() #misla da ja naprajme vizuelizacijava na ovie, pa na kraj pred treniranje da se skaliraat
```

```python
data_ml = data_dropped.copy() #ova da go koristime za ml, kje go enkodirame - so one hot na content type kolonata
#celta mi e da imame dve var razlicni za dl i za ml (nisto spec)
#prvo kje go napra enkodiranjeto
```

```python
dummies = pd.get_dummies(data_ml["content_type"], drop_first=False)
dummies = dummies.add_prefix("encoded_").astype(int)
data_ml = pd.concat([data_ml.drop(columns=["content_type"]), dummies], axis=1)
```
